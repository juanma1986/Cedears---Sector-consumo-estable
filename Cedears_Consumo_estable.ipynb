```python
stock = ["abev","agro","bg","cl","cost","deo","kmb","ko","mo","ntco","pep","pg","syy","tgt","ul","wmt"]
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Calculo del desvio estandar y volatilidad
```

```python
stocks = yf.download(stock)
data = stocks.loc[:, "Close"].copy()
data.pct_change()
```

```python
df = data.pct_change().dropna()
df.describe()
```

```python
df1 = df.describe().T.loc[:, ["mean", "std"]]
df1
```

```python
df1["mean"] = df1["mean"]*252
df1["std"] = df1["std"] * np.sqrt(252)
```

```python
df1.plot.scatter(x = "std", y = "mean", figsize = (12,8), s = 50, fontsize = 15)
for i in df1.index:
 plt.annotate(i, xy=(df1.loc[i, "std"]+0.002, df1.loc[i, "mean"]+0.002), size = 15)
```

**Conclusion**

Dentro del sector de los cedears defensivos y bajo el analisis de riesgo vs desempeño. Podemos concluir que el cedear de Adecoagro (AGRO) es mas riesgoso que el resto pero tiene un valor de retorno mayor.
